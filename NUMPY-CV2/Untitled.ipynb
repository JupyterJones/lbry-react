```python
#Numpy
#The NumPy library is the core library for scientific computing in Python. It provides a  high-performance multidimensional array object, and tools for working with these arrays.  
#Use the following improt convention:  

import numpy as np
#Numpy Arrays 
#Creating Arrays 
a = np.array([1,2,3])
print("a = np.array([1,2,3])\nresult:\n", a)
print("-----------------")
b = np.array([(1.5,2,3), (4,5,6)], dtype = float)
print("b = np.array([(1.5,2,3), (4,5,6)], dtype = float)\nresult:\n",b)
print("-----------------")
c = np.array([[(1.5,2,3), (4,5,6)],[(3,2,1), (4,5,6)]], dtype = float)
print ("c = np.array([[(1.5,2,3), (4,5,6)],[(3,2,1), (4,5,6)]], dtype = float)\nresult:\n",c)
print("-----------------")
d = np.zeros((3,4)) #Create an array of zeros
print ("np.zeros((3,4))\nresult:\n",d)
print("-----------------")
dd = np.ones((2,3,4),dtype=np.int16)
print ("np.ones((2,3,4),dtype=np.int16), #Create an array of ones\nresult:\n",dd)
print("-----------------")
g = np.random.random((2,2)) #Create an array with random values
print("np.random.random((2,2)) #Create an array with random values\nresult:\n",g)
print("-----------------")
h = np.empty((3,2)) #Create an empty array
print("h = np.empty((3,2)) #Create an empty array\nresult:\n", h)
ddd = np.arange(10,25,5)
print ("ddd = np.arange(10,25,5)  #Create an array of evenly spaced values (step value)\nresult:\n",ddd)
print("-----------------")
ls = np.linspace(0,2,9)
print ("ls = np.linspace(0,2,9)  #Create an array of evenlyspaced values (number of samples)\nresult:\n",ls)
print("-----------------")
e = np.full((2,2),7)
print ("e = np.full((2,2),7)  #Create a constant array\nresult:\n",e)
print("-----------------")
f = np.eye(2)
print ("f = np.eye(2)  #Create a 2X2 identity matrix\nresult:\n",f)
print("-----------------")
```
