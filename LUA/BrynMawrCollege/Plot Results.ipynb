```python
%matplotlib notebook
import matplotlib.pyplot as plt

index = list(range(5))
bar_width = 0.35

data = {}

with open("/home/dblank/polls/cs110/p1/1.results") as fp:
    line = fp.readline()
    while line:
        if ":" in line:
            user, choice = line.split(":")
            data[user.strip()] = int(choice.strip())
        line = fp.readline()

choices = [0] * 5
for datum in data.values():
    choices[datum] += 1
        
plt.bar([i for i in index], choices)
plt.xticks([i + bar_width for i in index], ('1', '2', '3', '4', '5'))
plt.show()
```
